```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('/Users/bell/Downloads/cruisecastlogs.csv')
```

```python
df_sub = df[['LatitudeDeg','LatitudeMin','LongitudeDeg','LongitudeMin','BottomDepth','StationNameID','UniqueCruiseID']]
```

```python
df_sub['Latitude'] = df.LatitudeDeg + df.LatitudeMin / 60
df_sub['Longitude'] = -1 * (df.LongitudeDeg + df.LongitudeMin / 60 )
```

```python
px.scatter_geo(df_sub[(df_sub['UniqueCruiseID'] == 'DY1504') | (df_sub['UniqueCruiseID'] == 'DY1606')| (df_sub['UniqueCruiseID'] == 'TN249')],'Latitude','Longitude', hover_name="StationNameID", hover_data=["BottomDepth"])
```
